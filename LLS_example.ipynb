```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,5)

def identity(x):
    return x
```

```python
xdata = np.linspace(1, 20, 100)
ydata = np.log(xdata)*np.random.uniform(5,10,len(xdata)) + 0.9*np.random.random(len(xdata))
xdata1 = xdata.reshape(len(xdata),1)
ydata1 = ydata.reshape(len(ydata),1)
xplt = np.arange(1,20, 0.1)

plt.subplot(1,2,1)
plt.plot(xdata,ydata,'o',mfc='none', mec='k',alpha=0.5)
plt.grid()
plt.show()
```

```python
def lls(data,b_vals,guess_func,xvals):
    A = np.hstack([guess_func(data),np.ones((len(data),1))])
    x_bar = np.linalg.solve(A.T.dot(A),A.T.dot(b_vals))
    y = float(x_bar[0])*guess_func(xvals) + float(x_bar[1])
    return y
```

```python
plt.subplot(1,2,1)
plt.plot(xdata,ydata,'o',mfc='none', mec='k',alpha=0.5)
plt.plot(xplt,lls(xdata1,ydata1,np.log,xplt),'r-.',label = 'Log',alpha=0.5)
plt.xlim(1,21)
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.plot(xdata,ydata,'o',mfc='none', mec='k',alpha=0.5)
plt.plot(xplt,lls(xdata1,ydata1,identity,xplt),'b--',label = 'Linear',alpha=0.5)
plt.xlim(1,21)
plt.grid()
plt.legend()

plt.show()
```
